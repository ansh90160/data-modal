```python
df['month'] = df['invoice_date'].dt.month
monthly_sales = df.groupby('month')['net_amount'].sum()
print(monthly_sales)
```

```python
#Description:
#df['month'] = df['invoice_date'].dt.month:

#Extracts the month from the invoice_date column (which should be in datetime format) and creates a new column called 'month' in the DataFrame.
#dt.month method extracts the month as an integer (1 for January, 2 for February, etc.).
#df.groupby('month')['net_amount'].sum():

#Groups the DataFrame by the month column.
#For each group (i.e., each month), it calculates the sum of the net_amount column, aggregating the total sales for each month.
#print(monthly_sales):

#Prints the result, which is a Series with the months as the index and the sum of net_amount for each month as the values
```

```python
sm = df.groupby(['sales_man_name']).agg({'market_area_id' :'count'})
```

```python
sm
```

```python
#Description:
#df.groupby(['sales_man_name']):

#Groups the DataFrame by the sales_man_name column, which means the data is organized by each salesperson.
#agg({'market_area_id': 'count'}):

#The .agg() function is used to apply an aggregation on the grouped data.
#For each salesperson (sales_man_name), it counts the number of non-null entries in the market_area_id column (using 'count').
#This provides the total count of market areas associated with each salesperson.
#Result (sm):

#The result is a new DataFrame or Series (sm) with sales_man_name as the index and the count of market_area_id as the values. This gives the number of market areas each salesperson is associated with.
```

```python
sm = df.groupby(['sales_man_name','brand']).agg({'brand' :'count'})
```

```python
sm
```

```python
#Description:
#df.groupby(['sales_man_name', 'brand']):

#Groups the DataFrame by two columns: sales_man_name and brand. This means the data is grouped by both salesperson and brand, creating unique combinations of each salesperson and brand.
#agg({'brand': 'count'}):

#The .agg() function applies an aggregation to the grouped data.
#For each combination of sales_man_name and brand, it counts the number of non-null entries in the brand column (using 'count').
#Essentially, this counts how many times each brand appears for each salesperson.
#Result (sm):

#The result is a new DataFrame or Series (sm) with a MultiIndex (comprising sales_man_name and brand) and the count of occurrences of each brand for each salesperson as the values.
```
